```python
#1
mycursor = cnx.cursor()
mycursor.execute("select property_type ,count(*), avg(Price),Square_Feet, Beds,Baths from properties group by property_type")
d1 = mycursor.fetchall()
for x in d1:
    print(x)
```

```python
plt.figure(figsize=(25,10))
plt.title('Count of the properties by each property type')
sns.barplot(x='Property_Type',y='No_of_Properties', data=abc)
```

```python
#2
mycursor2 = cnx.cursor()
mycursor2.execute("select s.SellerID, count(p.PropertyID) AS No_of_Houses_sold_by_eachseller From seller_dataset s join properties p On p.PropertyID =s.PropertyID group by s.SellerID")
p1 = mycursor2.fetchall()
for x in p1:
    print(x)
    
y1= pd.DataFrame(p1)

y1.columns = ['SellerID', 'No_of_Houses_sold_by_eachseller']

```
